### Project 2: Movie Data Cleaning 
#### Objective: Clean a movie dataset that contains duplicate entries and inconsistent values for a genre column. 
#### Technique: Use df.drop_duplicates() and string manipulation (.str.lower(), .str.replace(), etc.) to standardize the genre column.

##### Step 1: Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Step 2: Create a sample dataset

```python
data = {
    'Movie': ['Avatar', 'Titanic', 'Avatar', 'Inception', 'Titanic', 'Avengers', 'Avengers '],
    'Genre': ['Action', 'Romance ', ' ACTION', 'Sci-Fi', 'romance', 'Action ', 'action']
}

df = pd.DataFrame(data)

print("Original Dataset:")
print(df)
```

##### Step 3: Remove duplicates

```python
df = df.drop_duplicates()
```

##### Step 4: Clean and standardize the Genre column

```python
df['Genre'] = (
    df['Genre']
    .str.lower()          # Convert to lowercase
    .str.strip()          # Remove extra spaces
    .str.replace('-', ' ')  # Replace hyphens with spaces if any
)
```

##### Step 5: Also clean movie names (to fix trailing spaces)

```python
df['Movie'] = df['Movie'].str.strip()
```

##### Step 6: View cleaned dataset

```python
print("\n Cleaned Dataset:")
print(df)
```

##### Step 7: Visualize Genre Distribution

```python
# Count the number of movies per genre
genre_count = df['Genre'].value_counts()

# Create a bar plot 
plt.figure(figsize=(8, 5))
sns.barplot(x=genre_count.index, y=genre_count.values, hue=genre_count.index, palette='coolwarm', legend=False)

# Add titles and labels 
plt.title('Number of Movies per Genre', fontsize=14)
plt.xlabel('Genre')
plt.ylabel('Number of Movies')

# Add value labels on bars
for i, v in enumerate(genre_count.values):
    plt.text(i, v + 0.05, str(v), ha='center', fontweight='bold')

plt.tight_layout()
plt.show()
# enumerate() - op through a list (or array) and automatically gives both the index and the value at the same time.

# Counts how many movies belong to each genre
# Displays a clean bar chart
# Adds count labels above each bar for clarity
```
