### Project 1: Basic EDA on a Clean Dataset 
#### Objective: Perform a basic EDA on a well-structured dataset to understand its key characteristics. 

##### Step 1: Import required libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# pandas (pd) → helps store and explore data in tables (DataFrames).
# seaborn (sns) → used for stylish statistical graphs.
# matplotlib.pyplot (plt) → used for normal plotting and displaying graphs.
# sklearn.datasets.load_iris → loads the built-in Iris dataset easily.
```

##### Step 2: Load the Dataset

```python
iris_data = load_iris()
df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
df['species'] = iris_data.target
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# load_iris() loads a dataset containing:
# 4 numeric features:
# sepal length, sepal width, petal length, petal width
# target values (0,1,2) → represent 3 species of flowers.
# We convert it into a DataFrame (df) so it’s easier to handle.
# The column species originally has numbers (0, 1, 2).
# Using .map(), we replace them with names:
# 0 → setosa
# 1 → versicolor
# 2 → virginica
```

##### Step 3: Initial Inspection

```python
print(df.head())
print(df.tail())
print(df.info())
print(df.describe())

# .head() → shows the first 5 rows of data.
# .tail() → shows the last 5 rows of data.
# .info() → tells you:
# How many rows/columns
# Each column’s data type (float/int/object)
# If there are any missing values
# .describe() → gives summary statistics like mean, min, max, std, etc., for numeric columns.
# This step helps you understand the shape and health of the dataset.
```

##### Step 4: Explore the Categorical Column

```python
print(df['species'].value_counts())

# .value_counts() counts how many flowers belong to each species.
# It shows the dataset is perfectly balanced (50 of each type).
```

##### Step 5: Visualize Relationships

```python
# Pair Plot
sns.pairplot(df, hue='species', diag_kind='kde')
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()

# Pairplot → creates scatter plots for every feature combination.
# hue='species' → colors each species differently.
# diag_kind='kde' → draws smooth curves on the diagonal.
# Features relate to each other (correlations)
# Species cluster differently based on measurements
# Setting y=1.02 moves it slightly higher — so it doesn’t overlap with the plot or axis labels.

# Box Plot
plt.figure(figsize=(10, 6))
df.boxplot(by='species', figsize=(12, 8))
plt.suptitle("Box Plots of Iris Features by Species")
plt.show()

# A boxplot shows the spread and outliers for each feature.
# can easily spot:
# Median (center line)
# Quartiles (box range)
# Outliers (points outside whiskers)
# Comparing by species helps see how feature values differ between flower types
# Setting y=1.02 moves it slightly higher — so it doesn’t overlap with the plot or axis labels.
```
