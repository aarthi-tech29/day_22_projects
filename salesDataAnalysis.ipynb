### Project 1: Sales Data Analysis
#### Objective: Analyze a sales dataset to find the top-selling products and the average revenue per region. 
#### Technique: Use groupby to aggregate total sales and sort_values to rank products. Use groupby again to find the average revenue per region. 

##### Step 1: Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Step 2: Create or load a sales dataset

```python
data = {
    'Product': ['Laptop', 'Laptop', 'Mobile', 'Mobile', 'Tablet', 'Tablet', 'Headphones', 'Headphones'],
    'Region': ['North', 'South', 'East', 'West', 'North', 'East', 'South', 'West'],
    'Units_Sold': [120, 80, 150, 130, 90, 110, 200, 170],
    'Price_per_Unit': [800, 800, 500, 500, 300, 300, 100, 100]
}

df = pd.DataFrame(data)
```

##### Step 3: Add a new column for Total Sales (Revenue)

```python
df['Revenue'] = df['Units_Sold'] * df['Price_per_Unit']

print("----- Sales Dataset -----")
print(df)
```

##### Step 4: Find Top-Selling Products

```python
product_sales = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False).reset_index()
print("\n----- Total Sales by Product -----")
print(product_sales)
```

##### Step 5: Find Average Revenue per Region

```python
avg_region_revenue = df.groupby('Region')['Revenue'].mean().sort_values(ascending=False).reset_index()
print("\n----- Average Revenue per Region -----")
print(avg_region_revenue)
```

##### Step 6: Visualization Setup

```python
plt.figure(figsize=(14, 5))

plt.figure(figsize=(14, 5))

# Plot 1  Total Sales by Product
plt.subplot(1, 2, 1)
sns.barplot(data=product_sales, x='Product', y='Revenue', hue='Product', palette='Blues_d', legend=False)
plt.title('Total Sales by Product', fontsize=14)
plt.ylabel('Revenue')
plt.xlabel('Product')

# Plot 2  Average Revenue per Region
plt.subplot(1, 2, 2)
sns.barplot(data=avg_region_revenue, x='Region', y='Revenue', hue='Region', palette='Greens_d', legend=False)
plt.title('Average Revenue per Region', fontsize=14)
plt.ylabel('Average Revenue')
plt.xlabel('Region')

plt.suptitle('Sales Data Analysis  Insights Summary', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```
